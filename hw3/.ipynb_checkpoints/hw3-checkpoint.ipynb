**Exercise 1.** Request the value for the smoothed_cli signal from the fb-survey datasource on 2022-01-01 for ALL STATES.  Notice that the above is for counties and so you need to change this to states.  Do the same for smoothed_hh_cmnty_cli signal from the fb-survey data, and the confirmed_7dav_incidence_prop signal from the jhu-csse data source.  Create dataframes from these and join these 3 variables based on the state into a single dataframe.  Drop rows with missingness (there should be only about 5 of them dropped this way, corresponding to the US territories).  Finally, compute and output the spearman correlation between all variables in this dataframe (there should be 51 rows in the new dataset). 

````python
# define a function to generalize http get parameters
def params(data_source,signal):
    return dict( {
"data_source": data_source,
"signal": signal,
"time_type":"day",
"geo_type":"state",
"time_values":list([20220101,20220101]),
"geo_value":"*"
})

# get requests for each data source and signal and extract relevant columns
smoothed_cli = pd.DataFrame(requests.get("https://api.covidcast.cmu.edu/epidata/covidcast/",params = params("fb-survey","smoothed_cli")).json()["epidata"])[["geo_value","signal","value"]]
hh_cmnty = pd.DataFrame(requests.get("https://api.covidcast.cmu.edu/epidata/covidcast/",params = params("fb-survey","smoothed_hh_cmnty_cli")).json()["epidata"])[["geo_value","signal","value"]]
jhu_csse = pd.DataFrame(requests.get("https://api.covidcast.cmu.edu/epidata/covidcast/",params = params("jhu-csse","confirmed_7dav_incidence_prop")).json()["epidata"])[["geo_value","signal","value"]]
pd.merge(pd.merge(smoothed_cli,hh_cmnty, on="geo_value"), jhu_csse, on="geo_value")
````

````python
# merge all three dataframes together and calculate the spearman correlation 
corr_df = pd.merge(pd.merge(smoothed_cli,hh_cmnty, on="geo_value"), jhu_csse, on="geo_value").corr(method = "spearman")
# rename columns
corr_df.columns = ["smoothed_cli","hh_cmnty","jhu_csse"]
# rename rows
corr_df = corr_df.rename(index={"value_x": "smoothed_cli", "value_y": "hh_cmnty", "value": "jhu_csse"})

corr_df
````

**Exercise 2.**  Find the table displayed above, using the `find_all` method.  Then print the html for this table using `print(table_found.prettify())` if `table_found` is the BeautifulSoup object corresponding to the table of interest.

````python
# read local html file
with open('sp_500_market_cap.html','r') as file_ob:
    req_text = file_ob.read()
    
# initialize the BeautifulSoup object
soup = BeautifulSoup(req_text)
# find all tables
tables = soup.find_all("table")
# init an empty list
market_cap_tables = []

# iterate over all tables
for table in tables:
    # find the table head
    thead = table.find("thead")
    # find tables that contain Date in the header
    if thead and "Date" in thead.text:
        # append table html to a list
        market_cap_tables.append(table)
        
# print out the html for the 2 matching tables
for table in market_cap_tables:
    print(table.prettify())
````

**Exercise 3.** Find all of the table rows in this table that contain the market cap data.  Convert this into a pandas dataframe that contains a 'date' variable and 'market_cap' variable where 'date' is a datetime object, and 'market_cap' is a float.  Plot this timeseries. You can choose matplotlib, plotnine, or seaborn. The plot must have x-axis and y-axis and a title. 

````python
import itertools

# init another list for dfs 
dfs = []
# iterate over the 2 html tables
for table in market_cap_tables:
    # append the dfs from the html 
    dfs.append(pd.read_html(str(table)))
    
# flatten the nested lists and concat into one df
plot_df = pd.concat(list(itertools.chain(*dfs)))

plot_df.head()
````

````python
# change Date column to datetime
plot_df["Date"] = pd.to_datetime(plot_df["Date"])
# remove the "T" from the Value column
plot_df["Value"] = plot_df["Value"].str[:-1]
# change the Value column to a float
plot_df["Value"] = plot_df["Value"].astype(float)
# Change column name
plot_df = plot_df.rename(columns={"Value": "market_cap"})
# resent the df index
plot_df = plot_df.reset_index()
````

````python
import seaborn as sns
# create a line plot 
sns.lineplot(x = "Date",y = "market_cap",data = plot_df).set(title='S&P 500 Market Cap 2010-2023', ylabel = "Market Cap (T)", xlabel = "Year")
````
