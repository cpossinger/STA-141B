### Part 2: Read in the 'county_age_dist.csv' data and answer the next four questions.

__Exercise 2.1 (5 points).__ Cast the 'fips' column as a string such that it is a length 5 string of digits that starts with 0 if it is currently of length 4.  Yolo county for example is '06113'.  This is the age distribution of each county in the US (with territories), you may notice that the total population is unusually large, please ignore this.

```python
# define lambda function that prepends a string if its length is 4
prepend_zero = lambda s: "0" + s if len(s) == 4 else s
# read in dataa
counties = pd.read_csv("county_age_dist.csv")
# apply lambda function to the fips column
counties["fips"] = counties["fips"].apply(str).apply(prepend_zero)
```

__Exercise 2.2 (5 points).__ Create a population column which is the sum of the age counts for all age bins.  For each age bin create a column which is proportion of the age bin in the county populations, so for example '0-17_rat' is the ratio of '0-17' to the whole population for the county.  Print out the FIPS code for the county with the largest proportion of 0-17 year olds? What is the FIPS code for the county with the largest proportion of 85+ year olds?

```python
# define a function that obtains the row sum
def row_sum(row): 
    return row.drop("fips").sum() 
# apply the row_sum function along each row of the counties df
counties["population"] = counties.apply(row_sum, axis = 1)
```

```python
counties
```

```python
# define a function that calculates the row proportions 
def pop_age_ratio(row):
    row = row.drop("fips")
    return row / row["population"]

# create a new df that contains the row ratios
ratio_df = counties.apply(pop_age_ratio,axis = 1)
# Drop population column
ratio_df.drop(columns=['population'], inplace=True)
# add "_rat" to each column name ending
new_col_names = {col: col + "_rat" for col in ratio_df.columns}
# actually rename those columns
ratio_df.rename(columns=new_col_names, inplace=True)

# concatenate both dfs 
counties = pd.concat([counties,ratio_df],axis = 1)
```

```python
# obtain the largest 0-17_rat fips value
counties.loc[counties["0-17_rat"].idxmax(),"fips"]
```

```python
# obtain the largest 85+_rat fips value
counties.loc[counties["85+_rat"].idxmax(),"fips"]
```

__Exercise 2.3 (5 points).__ What is the overall proportion of each age bin within the entire US.  You need to find the total number of each age bin across all counties and take the ratio with the total population.

```python
# select relevent columns
us = counties.loc[:,["0-17","18-24","25-34","35-44","45-54","55-64","65-74","75-84","85+","population","fips"]]
# sum all columns and assign a new df
us = pd.DataFrame(us.sum()).T
# use the pop_age_ratio function and apply it to each row, drop the population column
us_ratio = us.apply(pop_age_ratio,axis = 1).drop(columns = "population",inplace = False)
us_ratio
```

__Exercise 2.4 (10 points).__ Create a new column that contains the mode of the age bins (the age bin with the largest proportion in the county).  Print out the frequencies of the modes different possible values, so that the frequency of '0-17' is the proportion of the counties that have '0-17' as the largest age bin for that county.

```python
# select relevent columns
county_ratio = counties.loc[:,["0-17","18-24","25-34","35-44","45-54","55-64","65-74","75-84","85+"]]
# obtain the largest county ratio index for each row
county_ratio = county_ratio.idxmax(axis = 1) 
# assign that series to a new df  
county_ratio = pd.DataFrame({"mode": county_ratio})

# concat the counties df with that ratio
counties = pd.concat([counties, county_ratio],axis = 1)
```

```python
# print the mode column and the value_counts associated with it
counties["mode"].value_counts(normalize = True)
```

__Exercise 3.2 (10 points)__ Drop any rows with missingness.  For each state calculate the proportion of each age bin as you did for the entire US in the question above.  What is the state with the most 0-17 proportion and the state with the most 85+ proportion?  Give the proportions.

### Part 4. Read in the 'time_series_covid19_deaths_US.csv' data and answer the next five questions

__Exercise 4.1 (5 points).__ This data is the Johns Hopkins official US Covid-19 death cumulative counts for each date and county.  Restructure the dataframe such that it is 'tidy' and there is a column for date, deaths, population, and fips.  (Hint: look at pandas melt function)

```python
# read in data
covid = pd.read_csv("time_series_covid19_deaths_US.csv")
```

```python
# get all date columns
date_cols = [col for col in covid.columns.to_list() if "/" in col]
# melt covid dataset into long format
covid =  covid.melt(id_vars = ["FIPS","Population","Admin2","Province_State"],value_vars = date_cols, var_name = "date",value_name = "deaths")
```

```python
covid
```

```python

# remove counties with 0 population
covid = covid.loc[covid["Population"] > 0]
```

__Exercise 4.2 (5 points).__ Which county (give county name and state) has the highest proportion of deaths by 2/25/21 to the Population?  Use the population in the covid deaths datasets.

```python
# create a filtering mask
mask = covid["date"] == "2/25/21"
# apply mask
covid_mask = covid.loc[mask,:] 
# create a new column prop 
covid_mask = covid_mask.assign(prop = covid_mask["deaths"] / covid_mask["Population"])
# reset index
covid_mask = covid_mask.reset_index()
```

```python
# get county with the highest proportion of deaths by 2/25/21
covid_mask.loc[covid_mask["prop"].idxmax(),["FIPS","Admin2","Province_State"]]
```

__Exercise 4.3 (10 points).__ Compute the incidence proportion (number of deaths each day per 100,000 population) for each county, you may save this as a new dataset.

```python
# create a new column that computes the incidence proportion
covid["incid_prop"] = (covid["deaths"] / covid["Population"]) * 100000
# create a copy df
covid_incid = covid
covid_incid
```

__Exercise 4.4 (15 points).__ Compute the incidence proportion per week (per 100,000 population), which is the sum of the incidences for all days in a week.  Hint: use Periods in Pandas.  Which is the week with the highest mean incidence proportions (where the mean is taken over all the counties)?

```python
# recast date column as a datetime column
covid_incid["date"] = pd.to_datetime(covid_incid["date"]) 
```

```python
# create a new column as a weekly period
covid_incid["week_period"] = covid_incid["date"].dt.to_period(freq = "W")
```

```python
# groupby the week period column
covid_incid = covid_incid.groupby(["week_period"])
```

```python
# find the week with the highest mean incidence proportions
covid_incid.mean().loc[:,"incid_prop"].idxmax()
```

__Exercise 4.5 (15 points).__ For each week (should be roughly 40 weeks), compute the spearman correlation between the proportion of seniors in the county and the incidence proportion for that county at that week (each county is considered to be a sample for computing the correlation).  Print all correlations below.  This should be a single number for each week.

```python
# calculate a new senior column
county_master["senior"] =  county_master["75-84"] + county_master["85+"]
# calculate a new senior proportion column
county_master["senior_rat"] = county_master["senior"] / county_master["population"]
```

```python
# select relevent columns and rename the FIPS column to match the other df
covid_corr = covid.loc[:,["FIPS","week_period","incid_prop"]].rename(columns = {"FIPS": "fips"},inplace = False)
# select relevent columns
county_corr = county_master[["fips","senior_rat"]]
# change the fips column to a float column
county_corr["fips"] = county_corr.loc[:,"fips"].astype(float)

# use a full join to merge both dfs on fips and then drop rows with missing values and reset the index
corr_df = pd.merge(covid_corr,county_corr, on = "fips", how = "outer").dropna().reset_index()
# filter out rows with an incid_prop equal to 0 and select relevent columns
corr_df = corr_df.loc[corr_df["incid_prop"] > 0,["week_period","incid_prop","senior_rat"]]
```

```python
corr_df
```

```python
# groupby week period column
corr_df = corr_df.groupby("week_period")
```

```python
# calculate spearman correlation
corr_df = corr_df.corr(method = "spearman")
```

```python
# drop rows with missing values and reset multiindex
corr_df = corr_df.dropna().reset_index()
```

```python
# extract only the relevent correlation value from the pairwise matrix and reset the index rename the corr column 
corr_df.loc[corr_df["level_1"] == "incid_prop",["week_period","senior_rat"]].reset_index(drop = True).rename(columns = {"senior_rat": "corr"})
```
